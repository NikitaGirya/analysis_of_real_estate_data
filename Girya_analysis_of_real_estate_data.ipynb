## Обработка пропущенных значений и приведение типов

Исходя из важности практически всех данных в текущем исследовании, предобработку проведем последовательно - по столбцам исходного датасета - для уменьшения вероятности пропустить скрытые артефакты.

Для начала создадим функцию `show_artifacts`, которая выводит количество отрицательных, нулевых и значений `NaN` по столбцу:

```python
# создание функции show_artifacts() с аргументом 'column'
def show_artifacts(column):
    '''
    Выводит количество отрицательных, нулевых 
    и значений 'NaN' по столбцу 'column'
    '''
    print('Отрицательных значений:', len(df[df[column] < 0]))
    print('Нулевых значений:', len(df[df[column] == 0]))
    print('Значений "NaN":', df[column].isna().sum())
```

Также напишем функцию `data_type_conversion` - приводит столбец к переданному в функцию типу данных и для проверки возвращает полученный тип столбца:

```python
# создание функции data_type_conversion() с аргументами 'columns и 'kind'
def data_type_conversion(column, kind):
    '''
    Приводит столбец 'column' к типу
    'kind' и возвращет новый тип столбца
    '''
    df[column] = df[column].astype(kind)
    
    print(df[column].dtype)
```

Функция `replacement_nan` поможет заменить `NaN` по столбцу на значение аргумента, переданного в функцию и для проверки вернет итоговое количество значений второго аргумента:

```python
# создание функции replacement_nan() с аргументами 'column' и 'value'
def replacement_nan(column, value):
    '''
    Заменяет значения 'Nan' в столбце 'column' на 'value' и
    возвращет итоговое количество значений второго аргумента
    '''
    df[column] = df[column].fillna(value)
    
    print(len(df[df[column] == value]))
```

Данное поле также без пропущенных значений. Проведем аналогичную первым трем столбцам проверку:

Возможно, нулевые значения числа комнат говорят о том, что квартира, к которой относится данное объявление, является студией. Проверим это:

```python
# получение числа записей по срезу "количество
# комнат == 0 и студия == 1" методом loc()
sum((df['rooms'] == 0) & (df['studio'] == 1))
```

Не все подобные предложения помечены как студии. 

Выведем медианную площадь студий с количеством комнат, равным нулю и медианную площадь квартир, отброшенных данным срезом:

```python
# получение медианного значения столбца 'total_area' методом
# median() по срезу "количество комнат == 0 и студия == 1"
# методом loc() с округлением до 2 знаков методом round()
df[(df['rooms'] == 0) & (df['studio'] == 1)]['total_area'].median().round(2)
```

```python
# получение медианного значения столбца 'total_area' методом
# median() по срезу "количество комнат == 0 и студия == 0"
# методом loc() с округлением до 2 знаков методом round()
df[(df['rooms'] == 0) & (df['studio'] == 0)]['total_area'].median().round(2)
```

Полученные значения практически равны, значит мы можем предположить, что оставшиеся после среза квартиры, все же являются студиями. 

Произведем соответствующую замену в столбце `studio`:

```python
# замена значений в столбце 'studio' на True методом mask() 
# по условию "количество комнат == 0 и студия == 0"
df['studio'] = df['studio'].mask((df['rooms'] == 0) & (df['studio'] == 0), True)

# проверка - получение числа записей по срезу "количество
# комнат == 0 и студия == 1" методом loc()
sum((df['rooms'] == 0) & (df['studio'] == 1))
```

Попробуем восстановить названия населенных пунктов. 

За первую "опорную" точку возьмем расстояние до аэропорта - в Санкт-Петербурге один действующий аэропорт гражданского назначения - "Пулково". Вероятно, что расстояние именно до него  имелось ввиду в значениях столбца `airports_nearest`. За вторую - расстояние до центра города - столбец `cityCenters_nearest`.

Создадим функцию `create_dict_of_lost_town`, которая принимает "столбец-ключ" с расстояниями (до аэропорта или центра города) и возвращает словарь со значениями - названиями населенных пунктов к которым эти расстояния относятся:

```python
# создание функции create_dict_of_lost_town() с аргументом 'column'
def create_dict_of_lost_town(column):
    '''
    Возвращает словарь с ключом - расстоянием до населенного
    пункта и значением - названием населенного пункта.
    '''
    # создание пустого словаря 'first_dict'
    first_dict = {}
    
    # заполнение словаря 'first_dict' сгруппированными по столбцу
    # 'locality_name' ключами и значениями - списками расстояний
    for key, val in df.groupby('locality_name')[column]:
        first_dict[key] = list(val)
        
    # "переворачивание" словаря 'first_dict' - теперь ключи - 
    # расстояния, значения - названия населенных пунктов
    # с сохранением в переменной 'second_dict'
    second_dict  = {x: old_keys for old_keys, old_val in first_dict.items() for x in old_val}
    
    # "очистка" словаря 'second_dict' от ключей 'NaN'
    # с сохранением в переменной 'clean_dict'
    clean_dict = {x: second_dict[x] for x in second_dict if not isnan(x)}
    
    # возврат словаря 'clean_dict'
    return clean_dict
```

Применим функцию последовательно к столбцам `airports_nearest` и `cityCenters_nearest`. С помощью полученных словарей заменим значения `Nan` в столбце `locality_name`:

```python
# замена пропущенных значений столбца 'locality_name' методом map() с использованием 
# словаря по столбцу 'airports_nearest' в качестве значений для подстановки
df.loc[df['locality_name'].isna(), 'locality_name'] = \
                                    df['airports_nearest'].map(create_dict_of_lost_town('airports_nearest'))
```

```python
# замена пропущенных значений столбца 'locality_name' методом map() с использованием 
# словаря по столбцу 'cityCenters_nearest' в качестве значений для подстановки
df.loc[df['locality_name'].isna(), 'locality_name'] = \
                                    df['cityCenters_nearest'].map(create_dict_of_lost_town('cityCenters_nearest'))
```

**Этаж квартиры: первый, последний, другой**

Создадим столбец `floor_type` с категориями этажности при помощи функции `create_floor_type`:

```python
# создание функции create_floor_type() с аргументом 'row'
def create_floor_type(row):
    ''' 
    Возвращает категорию этажности по номеру этажа сравнивая
    построчно значения столбца 'floor' со столбцом 'floors_total':
        - первый
        - последний
        - другой
        - 'unknown' при значении 'floors_total' == -1
    '''
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    elif (row['floor'] != 1) & (row['floors_total'] == -1):
        return 'unknown'
    return 'другой'
```

```python
# создание столбца 'floor_type' методом apply()
# с применением функции create_floor_type()
df['floor_type'] = df.apply(create_floor_type, axis=1)

# проверка - получение количества уникальных значений 
# столбца 'floor_type' методом value_counts()
df['floor_type'].value_counts()
```

С целью построения более наглядной гистограммы используем усеченную версию значений - по одному проценту с каждой стороны. Исходим из логики данных в таблице и предполагаем, что значения площади, цены, числа комнат, высоты потолков и т.д. в достаточной степени коррелируют между собой (но однозначно утверждать не можем - это необходимо проверить в последующем) - и, следовательно, выбросы по данным столбцам будут схожи, поэтому создадим функцию `trunc_column`, которая будет отсекать по 1% значений сверху и снизу:

```python
# создание функции trunc_column() с аргументом 'column'
def trunc_column(column):
    '''
    Возвращает усеченную по 1% сверху и снизу 
    версию столбца 'column' для более нагляд-
    ного визуального анализа значений 
    '''
    col = df[column]
    
    decile_1 = col.quantile(0.01)
    decile_9 = col.quantile(0.99)
    
    return df.query('@decile_1 < @col < @decile_9')[column]
```

Для автоматизации построения графиков создадим функцию `plot_hist_kde`, которая будет возвращать гистограмму и график плотности распределения по столбцу:

```python
# создание функции plot_estate_analysis() с аргументами (column, bins, color, axlabel, pltitle)
def plot_hist_kde(column, bins, color, axlabel, plttitle):
    '''
    Возвращает гистограмму и график плотности значений столбца 'column'
    с параметрами:
    - количество корзин - 'bins'
    - цвет - 'color'
    - подпись оси абсцисс - 'axlabel'
    - подпись графика - 'plttitle'
    '''
    # установление размера поля figsize=(13, 5)
    plt.figure(figsize=(13, 5))
    
    # вывод подписи графика plt.title=pltitle
    plt.title(plttitle)
    
    # применение функции trunc_column() к столбцу 'column' и построение 
    # гистограммы и графика плотности методом distplot() с количеством
    # корзин bins=bins цвета color=color, толщиной линии linewidth=1.5 
    # и соответствующей подписью axlabel=axlabel
    sns.distplot(column, bins=bins, color=color,\
                 kde_kws=dict(linewidth=1.5), axlabel=axlabel)
    
    # активация второстепенных делений осей
    plt.minorticks_on()
    # отображение основной сетки grid() 'major' и определение внешнего вида линий
    plt.grid(which='major', color = 'k', linewidth = 0.3)
    # отображение второстепенной сетки grid() 'minor' определение внешнего вида линий
    plt.grid(which='minor', color = 'k', linewidth = 0.3, linestyle = ':')
    
    # возврат изображения графика
    return plt.show()
```

### Обработка редких и выбивающихся значений

В данном разделе рассмотрим четыре наиболее важных для последующего исследования столбца: 
1. `last_price` — цена на момент снятия с публикации

Для автоматизации построения коробчатых диаграмм (***боксплотов***) создадим функцию `plot_boxplot`, которая будет возвращать коробчатую диаграмму по столбцу:

```python
# создание функции plot_boxplot() с аргументами (column, color, outlier, axlabel, pltitle)
def plot_boxplot(column, color, outlier, xlabel, plttitle):
    '''
    Возвращает коробчатую диаграмму (боксплот) значений 
    столбца 'column' с параметрами:
    - цвет - 'color'
    - отображение выбросов 'outlier' (True/False)
    - подпись оси абсцисс - 'axlabel'
    - подпись графика - 'plttitle'
    '''
    # установление размера поля figsize=(8, 3)
    plt.figure(figsize=(8, 3))
    
    # вывод подписи графика plt.title=pltitle
    plt.title(plttitle)

    # построение коробчатой диаграммы столбца 'column' методом 
    # boxplot() цвета color='color' и соответствующей подписью 
    # оси абсцисс xlabel=''
    sns.boxplot(column, color=color, showfliers=outlier).set(xlabel=xlabel)
    
    # активация второстепенных делений осей
    plt.minorticks_on()
    # отображение основной сетки grid() 'major' и определение внешнего вида линий
    plt.grid(which='major', color = 'k', linewidth = 0.3)
    # отображение второстепенной сетки grid() 'minor' определение внешнего вида линий
    plt.grid(which='minor', color = 'k', linewidth = 0.3, linestyle = ':')
    
    # возврат изображения графика
    return plt.show()
```

Предполагая довольно большое количество выбросов выше ***Q3 (верхней квартили) + 1.5 IQR (межквартильного размаха)***. Создадим функцию `cut_outlier`, принимающую на вход столбец и возвращающая два среза данных: меньше *Q1-1.5IQR* и больше *Q3+1.5IQR* по переданному в функцию столбцу: 

```python
# создание функции cut_outlier() с аргументом 
# 'column' и параметрами 'lower' и 'upper'
def cut_outlier(column, lower, upper):
    '''
    Возвращает два среза таблицы df: меньше Q1-1.5IQR
    и больше Q3+1.5IQR, которые являются записями, 
    содержащими выбросы по столбцу 'column'
    '''
    col = df[column]
    
    #расчет значений нижнего квартиля (Q1) и верхней квартили (Q3) 
    q1 = col.quantile(0.25)
    q3 = col.quantile(0.75)
    
    #расчет межквартильного размаха (IQR)
    iqr = q3 - q1
    
    #расчет значения нижнего 'lower_cut' и верхнего 'upper_cut' срезов
    lower_cut = q1 - (1.5 * iqr)
    upper_cut = q3 + (1.5 * iqr)
    
    #сохранение нижнего среза данных в 'lower_query' и верхнего в 'upper_query'
    lower_query = df.query('@col < @lower_cut')
    upper_query = df.query('@col > @upper_cut')
    
    # формирование условия для вывода срезов по результату работы функции
    if (lower==True) & (upper==False):
        return lower_query
    elif (lower==False) & (upper==True):
        return upper_query
    else:
        return lower_query, upper_query
```

**`last_price` — цена на момент снятия с публикации**

Построим боксплот стоимости жилья:

```python
# вызов функции plot_boxplot() с соответствующими аргументами
plot_boxplot(df['last_price']/1e6, '#c69f59', True, 'Стоимость квартир, млн.руб.',\
             'Коробчатая диаграмма стоимости квартир')
```

### Изучение влияния различных факторов на стоимость квартиры

Для определения влияния различных факторов на стоимость жилья и визуализации корреляции между ними создадим функции: 
* `plot_heatmap` - построения тепловой карты (***heatmap***) корреляции значений
* `plot_price_boxplot_caterory` - визуализации диаграмм размаха стоимости квартир по категориям в столбцах

```python
# создание функции plot_heatmap() с аргументами (data, colmap, plttitle, ax)
def plot_heatmap(data, colmap, plttitle, ax):
    '''
    Возвращает тепловую карту (heatmap) значений 
    таблицы 'data' с параметрами:
    - цветовая карта - 'colmap'
    - подпись графика - 'plttitle'
    - расположение - ax==False (для построения одиночного
    графика), ax[i] (для использования subplots)
    '''
    # если ax==False, строим одиночный график
    if ax==False:
        # вывод подписи графика
        plt.title(plttitle)
        
        # построение тепловой карты по столбцам таблицы 'data' методом heatmap() с отображением
        # коэфф. корреляции (annot=True, fmt='.2g'), регулировкой измерительной шкалы (vmin=-1, 
        # vmax=1, center= 0, cbar_kws= {'orientation': 'horizontal'}), параметрами линий 
        # (linewidths=1, linecolor='white') и цветовой гаммой cmap=colmap
        sns.heatmap(data.corr(), annot=True, fmt='.2g', vmin=-1, vmax=1, center= 0,\
                    cbar_kws={'orientation': 'horizontal'}, linewidths=1, linecolor='white', cmap=colmap)
        
        # вывод изображения графика
        return plt.show()
    
    # в противном случае используем переданный в функцию номер графика 
    else:
        # построение графика для использования subplots
        ax.set_title(plttitle)
        sns.heatmap(data.corr(), annot=True, fmt='.2g', vmin=-1, vmax=1, center= 0,\
                    cbar_kws={'orientation': 'horizontal'}, linewidths=1, linecolor='white', cmap=colmap, ax=ax)
```

```python
# создание функции plot_price_boxplot_caterory() с аргументами 
# (data, x_column, y_column, xlabel, plttitle, colmap)
def plot_price_boxplot_caterory(data, x_column, y_column, x_label, plttitle, colmap, ax):
    '''
    Возвращает коробчатую диаграмму (боксплот) стоимости квартир 
    таблицы 'data' по столбцу 'column' с параметрами:
    - ось абсцисс - 'x_column'
    - подпись оси абсцисс - 'x_label'
    - ось ординат - 'y_column'
    - подпись графика - 'plttitle'
    - цветовая карта - 'colmap'
    - расположение - ax==False (для построения одиночного
    графика), ax[i] (для использования subplots)
    '''
    # если ax==False, строим одиночный график
    if ax==False:
        # вывод подписи графика
        plt.title(plttitle)
        
        # построение нормализованной (showfliers=False) коробчатой диаграммы столбца 
        # 'column' методом boxplot() в цветовой гамме cmap=colmap с соответствующими  
        # подписями осей (xlabel='', ylabel='')
        sns.boxplot(data=data, x=x_column, y=y_column, showfliers=False,palette=colmap).\
                    set(xlabel=x_label, ylabel='Стоимость квартир, млн.руб.')
        
        # активация второстепенных делений осей
        plt.minorticks_on()
        # отображение основной сетки grid() 'major' и определение внешнего вида линий
        plt.grid(which='major', color = 'k', linewidth = 0.2)
        # отображение второстепенной сетки grid() 'minor' определение внешнего вида линий
        plt.grid(which='minor', color = 'k', linewidth = 0.3, linestyle = ':')
        
        # вывод изображения графика
        return plt.show()
    
    # в противном случае используем переданный в функцию номер графика 
    else:
        # построение графика для использования subplots
        ax.set_title(plttitle)
        sns.boxplot(data=data, x=x_column, y=y_column, showfliers=False, palette=colmap, ax=ax).\
                    set(xlabel=x_label, ylabel='Стоимость квартир, млн.руб.')
        ax.minorticks_on()
        ax.grid(which='major', color = 'k', linewidth = 0.2)
        ax.grid(which='minor', color = 'k', linewidth = 0.3, linestyle = ':')
```

Визуализируем влияние на стоимость квартиры `last_price` факторов площади `total_area`, числа комнат `rooms` и удаленности от центра `cityCenters_nearest` для чего построим тепловую карту корреляции значений по данным столбцам:

```python
# установление размера поля
plt.figure(figsize=(8, 6))

# вызов функции plot_heatmap() с соответствующими аргументами
plot_heatmap(df_mod[['last_price', 'total_area', 'rooms', 'cityCenters_nearest']],\
             'BrBG', 'Тепловая карта факторов стоимости квартир', False)
```

Умеренная отрицательная связь с удаленностью от центра (*r = -0.31*) свидетельствует о близком к незначительному уменьшении цены при удаленности от центра и наоборот.

### Анализ влияния различных факторов на стоимость недвижимости в центре города

Для начала определим, какой именно в числовом выражении радиус, собственно, является центром города.

Создадим новый столбец `cityCenters_nearest_km`, в который запишем полученные делением на `1000` значения расстояния до центра города `cityCenters_nearest` в километрах, округленные до целых. Также создадим поле `avg_price_per_km` - средняя цена за километр, где сохраним результат деления столбца `last_price` на `cityCenters_nearest_km`.

```python
# преобразование столбца 'cityCenters_nearest' в километры
# делением на 1000 (1e3), и округление до целых значений 
# сохранением частного в столбец 'cityCenters_nearest_km'
df_mod['cityCenters_nearest_km'] = (df_mod['cityCenters_nearest']/1e3).round(0)

# вычисление средней цены для каждого километра как частного
# столбцов 'last_price' и 'cityCenters_nearest_km' с сохранением
# в столбец 'avg_price_per_km'
df_mod['avg_price_per_km'] = df_mod['last_price']/df_mod['cityCenters_nearest_km']
```

Построим график для визуального определения центральной зоны города, по оси ординат которого - значения столбца `avg_price_per_km`, а ось абсцисс - значения `cityCenters_nearest_km`:

```python
# установление размера графика figsize=(12, 6)
plt.figure(figsize=(12, 6))
# установление лимита по y=[-1, 50]
plt.ylim([-1, 50])

# вывод подписи графика
plt.title('Диаграмма разброса значений средней цены за километр по удаленности от центра')

# построение диаграммы разброса значений 'avg_price_per_km' (с делением столбца на 
# 1 000 000 (1e6) для отображения оси абсцисс в млн.руб.) по столбцу 'cityCenters_nearest_km'
# методом scatterplot() цвета color='seagreen' с соответствующими подписями осей (xlabel='', ylabel='')
sns.scatterplot(data=df_mod, x=df_mod['cityCenters_nearest_km'],\
            y=df_mod['avg_price_per_km']/1e6, color='seagreen').\
            set(xlabel='Удаленность от центра, км.', ylabel='Средняя цена за километр, млн.руб.')

# активация второстепенных делений осей
plt.minorticks_on()
# отображение основной сетки grid() 'major' и определение внешнего вида линий
plt.grid(which='major', color = 'k', linewidth = 0.2)
# отображение второстепенной сетки grid() 'minor' определение внешнего вида линий
plt.grid(which='minor', color = 'k', linewidth = 0.3, linestyle = ':')
```

График резко убывает на значении в ~`7` км. Этот радиус и возьмем для определения центральной зоны города. Сохраним срез таблицы `df_mod` с записями, в которых расстояние до центра меньше 7 км. в таблицу **`df_mod_center`**:

```python
# получение среза таблицы df_mod по условию "расстояние до центра 
# города <= 7" методом query() с сохранением в переменную 'df_mod_center'
df_mod_center = df_mod.query('cityCenters_nearest_km <= 7').reset_index(drop=True)

# получение размера таблицы df_mod_center методом shape
df_mod_center.shape
```

* продолжительность размещения объявления была восстановлена полностью на основании предположения о действующем объявлении на момент получения датафрейма и расчета продолжительности публикации как разности даты выгрузки данных и даты размещения объявления 

На ***втором этапе*** исследования были добавлены в таблицу такие важные для анализа параметры, как:
* цена квадратного метра , среднее значение которой позволило составить рейтинг населенных пунктов по стоимости жилья
* день недели, меясяц и год публикации дали возможность проанализировать влияние даты размещения на стоимость квартиры
* категория этажности квартиры: первый, последний, другой -  разделение по которым также указало на возможную зависимость цены жилья от его этажности 
* соотношение жилой и общей площади и отношение площади кухни к общей являющиеся показателями полезного пространства квартиры

По итогу изучения в ходе ***третьего этапа*** всевозможных факторов, влияющих на стоимость квартиры, можем утверждать, что:
* **наиболее популярны квартиры площадью от 30 до 50 квадратных метров**. Основные пики значений - 30 и 40 м². Менее востребованы квартиры от 50 до 75 м² и далее с возрастанием квадратуры жилья наблюдаем убывание числа объявлений
* **самый востребованный ценовой сегмент - от 3 до 5 млн.руб. с пиком в районе 4 млн.руб.** Также, как и в случае с площадью, далее идут довольно редкие значения с большей ценой
* распределение количества комнат указывает, что **наиболее популярные - однокомнатные и двухкомнатные квартиры**, за ними - трехкомнатные. Минимальной привлекательностью обладают студии и квартиры больше пяти комнат
* **стандартная высота потолков в квартирах - от 2.5 до 3 метров**, определенная часть присутствует в диапазоне 3-4 метров
* **обычно продажа занимает около 2.5 месяцев** - вероятно, данные квартиры попадают в наиболее востребованный сегмент жилья, а, как было выяснено на предыдущем этапе, - это 1-2-комнатные квартиры за 3-5 млн.руб.
* **очень быстрые сделки происходят уже в день публикации** - это могут быть необычайно интересные предложения по низкой (ввиду разных обстоятельств) цене с иными благоприятствующими факторами: небольшой удаленностью от центра, оптимальным количеством комнат, высокими потолками и т.д. Также возможно, что данные квартиры выкупают сами риэлторы с целью дальнейшей перепродажи
* **необычно долгие объявления размещены больше четырех лет**, что, скорее всего, связано с различными причинами: слишком высокая стоимость или площадь квартиры, удаленность от центра города, нахождение в неблагоприятном районе и т.д., а также их совокупностью

По итогу **обработки редких и выбивающихся и некорретных значений** имеем:
* **выбросы по стоимости обусловлены элитностью жилья**, о чем свидетельствует редкость значений, большая площадь помещений в данных записях, соответствующее площади количество комнат, а также высокая длительность продажи подобных предложений
* **записи с выбросами по площади - нестандартные квартиры** с количеством комнат 5 и более, довольно высокой стоимостью по сравнению с типовым жильем, а также длительной публикацией, что говорит о невысокой востребованности данных предложений
* **в столбце со значениями высоты потолков обнаружены и обработаны явные ошибки в данных** связанные, вероятно, с человеческим фактором
* **большое количество комнат соответствует большой площади квартир**, значения стоимости в данных записях в достаточной степени высоки. Также часть этих предложений значительно удалена от центра, и, скорее всего, является элитным загородным жильем

**Наиболее дорогим городом** для покупки квартиры из представленных в исследовании вполне логично **является Санкт-Петербург** - средняя цена за квадратный метр тут разительно выше, чем в других населенных пунктах. **Дешевле всего** из первой десятки по количеству объявлений **жилье обойдется в Выборге** - почти в два раза дешевле Санкт-Петербурга. 

По итогам **анализа влияния факторов площади, количества комнат, удаленности от центра и даты размещения на стоимость квартиры** делаем следующие выводы:
* **стоимость квартиры сильно зависит от ее площади**, исходя из этого, также **прослеживается заметная зависимость от количества комнат**. Умеренная отрицательная связь с удаленностью от центра свидетельствует о близком к незначительному взаимном влиянии друг на друга данных параметров
* **медианная стоимость квартир на первых этажах несколько ниже**, чем по другим категориями этажности. Размах значений цены также выше по отличным от первого этажах, что свидетельствует о невысоком предпочтении покупателями первых этажей, выраженную в денежном эквиваленте
* **стоимость квартиры не всегда зависит от дня недели размещения** - медианная цена абсолютно незначительно выше в первой половине недели и падает к ее концу, самые дорогие предложения по статистике размещаются в четверг, самые дешевые - в среду, минимальный разброс по цене - в воскресенье
* **стоимость жилья зависит от месяца публикации объявления** - в апреле заметен рост стоимости предложений по квартирам, медианная цена также несколько выше, чем по остальным месяцам, в июне рынок недвижимости просаживается - разброс предложений и медианная цена ниже, самые дешевые квартиры публикуются в марте, дорогие - в сентябре
* **наблюдается спад рынка жилой недвижимости после 2014 года** - вероятно, ввиду повышения курса доллара и некоторых политических причин. После некоторой стагнации в 2016-2018 годах, цены снова поднимаются в 2019 году

Проанализировав **влияние различных факторов на стоимость недвижимости в центральной зоне** можно утверждать следующее:
* **наблюдается уже не такая высокая зависимость стоимости квартиры от площади, но она все еще высока**, а также прослеживается умеренная зависимость от количества комнат. Слабая отрицательная связь с удаленностью от центра свидетельствует об отсутствии корреляции, а это, в свою очередь, о том, что практически все самые дорогие предложения по недвижимости сконцентрированы в центре города
* **различия в стоимости квартир по категориям этажности в центре города намного выше**, чем в общем по всем предложениям. Соотношения ценового размаха в случае с центральной областью города также выше - значительно меньшее количество предложений по первым этажам свидетельствуют о нежелании людей приобретать подобную недвижимость 
* **зависимость стоимости квартиры от дня недели размещения сильнее прослеживается по центральной области города**. Медианная цена в целом равномерно распределена по дням недели, но с просадкой в субботу. Самые дорогие предложения в центре размещаются в понедельник, самые дешевые - во вторник. Также во вторник наблюдается и минимальный разброс по цене
* **также по центру города ощутимее влияние месяца размещения объявления**: медианная цена квартир выше в сентябре, в феврале и ноябре рынок недвижимости на подъеме - разброс предложений и медианная цена несколько выше, чем по другим месяцам. С февраля по май наблюдается спад медианной цены и размаха стоимости жилья, самые дешевые квартиры публикуются в декабре, дорогие - в феврале
* также, как и в случае со всеми предложениями, **по центральной области наблюдается резкий спад рынка жилой недвижимости после 2014 года**. В 2019 году квартиры снова дорожают, но до докризисных показателей все еще очень далеко
